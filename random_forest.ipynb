```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import seaborn as sns
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df
```

```python
df.drop('User ID',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
df['Gender'] = LabelEncoder().fit_transform(df['Gender'])
```

```python
df
```

```python
sns.pairplot(df,hue='Purchased')
```

```python
x = df.drop('Purchased',axis=1)
y = df['Purchased']
```

```python
from sklearn.preprocessing import StandardScaler
x = StandardScaler().fit_transform(x)
```

```python
x.shape
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier()
clf.fit(x_train,y_train)
train_preds = clf.predict(x_train)
test_preds = clf.predict(x_test)
print('train acc' , accuracy_score(y_train,train_preds))
print('test acc' , accuracy_score(y_test,test_preds))
```

```python
for md in range(1,13):
    clf = DecisionTreeClassifier(max_depth=md,random_state=10)
    clf.fit(x_train,y_train)
    train_preds = clf.predict(x_train)
    test_preds = clf.predict(x_test)
    print('md',md)
    print('train acc' , accuracy_score(y_train,train_preds))
    print('test acc' , accuracy_score(y_test,test_preds))
    print('----------------------------------------------')
```

```python
for md in range(1,13):
    clf = RandomForestClassifier(max_depth=md,random_state=10)
    clf.fit(x_train,y_train)
    train_preds = clf.predict(x_train)
    test_preds = clf.predict(x_test)
    print('md',md)
    print('train acc' , accuracy_score(y_train,train_preds))
    print('test acc' , accuracy_score(y_test,test_preds))
    print('----------------------------------------------')
```

```python
for n in [1,5,10,20,30,50,70,100,200,300]:
    clf = RandomForestClassifier(n_estimators=n,max_depth=4,random_state=10)
    clf.fit(x_train,y_train)
    train_preds = clf.predict(x_train)
    test_preds = clf.predict(x_test)
    print('n',n)
    print('train acc' , accuracy_score(y_train,train_preds))
    print('test acc' , accuracy_score(y_test,test_preds))
    print('----------------------------------------------')
```

```python
import pylab as pl
min_age , max_age = x[:,1].min() , x[:,1].max()
min_es , max_es = x[:,2].min() , x[:,2].max()
age , es = np.meshgrid(np.arange(min_age,max_age,0.02),np.arange(min_es,max_es,0.02))
```

```python
pl.figure(figsize=(15,30))
i = 1
for md in range(1,13):
    clf = DecisionTreeClassifier(max_depth=md,random_state=10)
    clf.fit(x_train[:,1:],y_train)
    train_preds = clf.predict(x_train[:,1:])
    test_preds = clf.predict(x_test[:,1:])
    print('md',md)
    print('training accuracy',accuracy_score(train_preds,y_train))
    print('test accuracy',accuracy_score(test_preds,y_test))

    Z = clf.predict(np.c_[age.ravel(),es.ravel()])
    Z = Z.reshape(age.shape)

    pl.subplot(12,2,i)
    pl.set_cmap(pl.cm.cividis)
    pl.contourf(age,es,Z)
    pl.axis('tight')
    pl.scatter(x_train[:,1],x_train[:,2],c=y_train)
    pl.title(f'training with md {md}')

    i+=1
    pl.subplot(12,2,i)
    pl.set_cmap(pl.cm.cividis)
    pl.contourf(age,es,Z)
    pl.axis('tight')
    pl.scatter(x_test[:,1],x_test[:,2],c=y_test)
    pl.title(f'testing with md {md}')
    pl.axis('tight')
    i+=1

pl.show()
```

```python
pl.figure(figsize=(15,30))
i = 1
for md in range(1,13):
    clf = RandomForestClassifier(n_estimators=100,max_depth=md,random_state=10)
    clf.fit(x_train[:,1:],y_train)
    train_preds = clf.predict(x_train[:,1:])
    test_preds = clf.predict(x_test[:,1:])
    print('md',md)
    print('training accuracy',accuracy_score(train_preds,y_train))
    print('test accuracy',accuracy_score(test_preds,y_test))

    Z = clf.predict(np.c_[age.ravel(),es.ravel()])
    Z = Z.reshape(age.shape)

    pl.subplot(12,2,i)
    pl.set_cmap(pl.cm.cividis)
    pl.contourf(age,es,Z)
    pl.axis('tight')
    pl.scatter(x_train[:,1],x_train[:,2],c=y_train)
    pl.title(f'training with md {md}')

    i+=1
    pl.subplot(12,2,i)
    pl.set_cmap(pl.cm.cividis)
    pl.contourf(age,es,Z)
    pl.axis('tight')
    pl.scatter(x_test[:,1],x_test[:,2],c=y_test)
    pl.title(f'testing with md {md}')
    pl.axis('tight')
    i+=1

pl.show()
```
